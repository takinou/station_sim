```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Nagoya_Station_data = pd.read_csv("./datas/Nagoya_Station_data.csv")
Nagoya_Station_data
```

```python
#jrなどの利用者数を合算する為にgroupbyする
datas = Nagoya_Station_data.groupby(['station_name','year'])[['josha_num','lines']].sum()

print(datas)
```

```python
#横軸 駅数、縦軸　年間乗車数


datas.plot(x='lines',y='josha_num', kind='scatter')
```

```python
datas.to_csv('output/1_sum_datas.csv')
```

```python
datas.describe()
```

```python
#相関係数
np.corrcoef(datas['lines'], datas['josha_num'])
```

```python
X=datas.lines.values
X
# X = np.vstack(datas.sum().lines)
# X = np.array([[value, 1] for value in X])
```

```python
#Xと同じ長さの列を作成して、1埋めする
A = np.vstack([X, np.ones(len(X))]).T
A
```

```python
Y = datas.josha_num.values
#Y=np.c_[Y]
Y
```

```python
#横軸 駅数、縦軸　年間乗車数

plt.plot(X,Y,'o')
```

```python
#傾きa, 切片bを求める

a,b = np.linalg.lstsq(A,Y,rcond=None)[0]
```

```python
print(a,b)
```

```python
#推定Yのplot
plt.plot(X, a*X + b,'o')
```

```python
#決定係数を手計算する

#残差計算
child = ((datas['josha_num'] - (datas['lines']*a + b))**2).sum()
child
```

```python
#分母の計算
mother = ((datas['josha_num'] - datas['josha_num'].mean())**2).sum()
mother
```

```python
#決定係数
R = 1 - child / mother
R
```
