```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Nagoya_Station_data = pd.read_excel("./datas/Nagoya_Station_data.xlsx")
Nagoya_Station_data
```

```python
#jrなどの利用者数を合算する為にgroupbyする
datas = Nagoya_Station_data.groupby(['station_name','year'])[['josha_num','lines']].sum()

print(datas)
```

```python
################ここ以降は、josha数の対数でとって計算してみる。

#非0のデータのみにしておく
#非0となる配列
datas['josha_num'].nonzero()[0]
```

```python
#非0のみのデータ
nz_datas=datas[['josha_num','lines']].iloc[datas['josha_num'].nonzero()[0]]
nz_datas
```

```python
nz_datas['log_josha_num']=np.log(nz_datas['josha_num'])

nz_datas
```

```python
nz_datas.plot(x='lines',y='log_josha_num', kind='scatter')
```

```python
#相関係数
#logを取らない場合の方が相関係数が高い。。。
np.corrcoef(nz_datas['lines'], nz_datas['log_josha_num'])
```

```python
X = nz_datas.lines.values

A = np.vstack([X, np.ones(len(X))]).T

Y = nz_datas.log_josha_num.values

plt.plot(X,Y,'o')
```

```python
a,b = np.linalg.lstsq(A,Y,rcond=None)[0]

print(a,b)
```

```python
#決定係数を手計算する

#残差計算
child = ((nz_datas['log_josha_num'] - (nz_datas['lines']*a + b))**2).sum()
child
```

```python
#分母の計算
mother = ((nz_datas['log_josha_num'] - nz_datas['log_josha_num'].mean())**2).sum()
mother
```

```python
#決定係数
#0.39なので、対数取らないで計算した方が良さそう。。。。

R = 1 - child / mother
R
```
