```python
multi_datas = Nagoya_Station_data.groupby(['station_name','year'])[['josha_num','lines','ward_office']]

print(multi_datas.sum())
```

```python
multi_datas.sum().plot(x='ward_office',y='josha_num', kind='scatter')
```

```python
#相関係数
np.corrcoef(multi_datas.sum()['ward_office'], datas.sum()['josha_num'])
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
```

```python
# 説明変数に "josha_num" を利用
datas_except_josha = multi_datas.sum().drop("josha_num", axis=1)
datas_except_josha 
```

```python
X = datas_except_josha.as_matrix()
X
```

```python
# 目的変数に "josha_num" を利用
Y = multi_datas.sum()['josha_num'].as_matrix()
Y
```

```python
# 予測モデルを作成
clf.fit(X, Y)
```

```python
# 偏回帰係数
print(pd.DataFrame({"Name":datas_except_josha.columns,
                    "Coefficients":clf.coef_}).sort_values(by='Coefficients') )
```

```python
# 切片 (誤差)
print(clf.intercept_)
```
