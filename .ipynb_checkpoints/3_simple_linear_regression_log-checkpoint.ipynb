```python
#名古屋駅、栄、金山など、極端に乗客数が多い所の値が強く出てしまっているように感じるので、logで表記して改善するかを確認する

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Nagoya_Station_data = pd.read_excel("./datas/Nagoya_Station_data(josha).xlsx")

print(Nagoya_Station_data)
```

```python
#利用線数が0 つまりまだ駅ができていない場所を取り除く
Nagoya_Station_data=Nagoya_Station_data[Nagoya_Station_data['lines']!=0]

```

```python
#jrなどの利用者数を合算する為にgroupbyする
datas = Nagoya_Station_data.groupby(['station_name','year'])[['josha_num','lines']].sum()

print(datas)
```

```python
################ここ以降は、josha数の対数でとって計算してみる。
datas['log_josha_num']=np.log(datas['josha_num'])
datas
```

```python
datas.plot(x='lines',y='log_josha_num', kind='scatter')
```

```python
#相関係数
#logを取らない場合の方が相関係数が高い。。。
np.corrcoef(datas['lines'], datas['log_josha_num'])
```

```python
X = datas.lines.values

A = np.vstack([X, np.ones(len(X))]).T

Y = datas.log_josha_num.values

#plt.plot(X,Y,'o')
```

```python
a,b = np.linalg.lstsq(A,Y,rcond=None)[0]

print(a,b)
```

```python
#推定Yのplot
X2 = np.arange(0, 10, 0.01)
plt.plot(X2, a*X2 + b)

plt.plot(X,Y,'ro')
```

```python
#決定係数を手計算する

#残差計算
child = ((datas['log_josha_num'] - (datas['lines']*a + b))**2).sum()
child
```

```python
#分母の計算
mother = ((datas['log_josha_num'] - datas['log_josha_num'].mean())**2).sum()
mother
```

```python
#決定係数
#0.39なので、対数取らないで計算した方が良さそう。。。。

R = 1 - child / mother
R
```
